```python
searchURL='http://data.bioontology.org/search'
```

```python
# reading the matrix and get the names of the features
merged = pd.read_table("/Users/jerome/Projects/openfda/merged_matrix.txt")
```

```python
colnames = list(merged.columns.values)
```

```python
len(colnames)
```

```python
# get the event list 
term_list = colnames[4:-1]
```

```python
len(term_list)
```

```python
# build the term dict
term_dict = {}
ct = 0
for key_term in term_list:
    ct += 1
    print("------------------ Processing Keyword " + str(ct) + " ------------------")
    term_dict[key_term] = []
    params1 = {'apikey':APIKEY,'ontologies':ONT,'q':key_term,'include':{'prefLabel', 'synonym', 'definition', 'notation', 'cui'}}
    req = requests.get(searchURL, params=params1)
    data = req.json()
    for collection in data["collection"]:
        try:
            term_dict[key_term] = term_dict[key_term] + (collection["synonym"])
            #print(collection["synonym"])
        except KeyError:
            pass
    
    
```
