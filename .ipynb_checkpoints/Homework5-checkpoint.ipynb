### Question 1

### a

$$T(n) \geq (1-p)T(1) + p\frac{T(1)}{n}$$

$$speedup(p) = \frac{T(1)}{T(n)}$$

$$speedup(p) \leq \frac{T(1)}{(1-p)T(1) + p\frac{T(1)}{n}}$$

$$speedup(p) \leq \frac{1}{(1-p) + \frac{p}{n}}$$

### b

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# speed up function
def speedup(p, n):
    return(1/((1-p)+p/n))
```

```python
# plotting function
def plot_speedup(p):
    max_num = 16385
    n_vec = np.arange(1, max_num)
    s = [speedup(p, x) for x in n_vec]
    ax = plt.gca()
    ax.scatter(n_vec, s)
    ax.set_xscale('log')
    plt.xlabel('log of the number of the preocessors')
    plt.ylabel('Speed up')
    plt.title('Speed up for p = ' + str(p))
```

```python
# store the values
def cal_speedup(p):
    max_num = 16385
    n_vec = np.arange(1, max_num)
    s = [speedup(p, x) for x in n_vec]
    return(s)
```

### c

```python
# p = 0.5
plot_speedup(0.5)
```

```python
# p = 0.67
plot_speedup(0.67)
```

```python
# p = 0.95
plot_speedup(0.95)
```

```python
# p = 0.99
plot_speedup(0.99)
```

```python
n_vec = list(np.arange(1, 16385))
s1 = cal_speedup(0.5)
s2 = cal_speedup(0.67)
s3 = cal_speedup(0.95)
s4 = cal_speedup(0.99)
ax = plt.gca()
ax.scatter(n_vec, s1, s = 3)
ax.scatter(n_vec, s2, s = 3)
ax.scatter(n_vec, s3, s = 3)
ax.scatter(n_vec, s4, s = 3)
ax.set_xscale('log')
plt.xlabel('log of the number of the preocessors')
plt.ylabel('Speed up')
```
