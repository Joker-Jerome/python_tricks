## Multiprocessing and Pipe 

```python
import os, random
from multiprocessing import Process, Pipe
```

```python
# producer task
def producer_task(conn):
    value = random.randint(1, 10)
    conn.send(value)
    print('Value [%d] sent by PID [%d]' % (value, os.getpid()))
    conn.close()
```

The key point of this function is called conn.send(value), which uses a
connection object generated by Pipe in the flux of the main program that has been
sent as an argument to the function.
